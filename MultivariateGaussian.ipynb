## Generating a 3D plot of the density of the Multivariate Standard Normal Distribution in two variables.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [15,7]
```

```python
x1 = np.linspace(-4, 4, 1000)
x2 = np.linspace(-4, 4, 1000)
x1_dash, x2_dash = np.meshgrid(x1, x2) # 2 Matrices of x1 & x2 coordinates in our grid.
f = 1/(2*np.pi)*np.exp(-(x1_dash**2 + x2_dash**2)/2) # Elementwise multiplication to compute f.
ax = plt.axes(projection='3d')
ax.plot_surface(x1_dash,x2_dash,f, cmap='plasma'); # 3-D Plot.
```

## Scatter Plot of Multivariate Standard Normal Distribution in two variables.

```python
x1 = np.random.randn(1000) # 1000 Standard Normally distributed points.
x2= np.random.randn(1000)
ax = plt.gca()
ax.set_aspect(1)
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.scatter(x1,x2,c='green');
```

## Contour plot of a normal distribution.

```python
x1 = np.linspace(-4, 4, 1000)
x2 = np.linspace(-4, 4, 1000)
x1_dash, x2_dash = np.meshgrid(x1, x2)
f = 1/(2*np.pi)*np.exp(-(x1_dash**2 + x2_dash**2)/2)
ax = plt.gca()
ax.set_aspect(1)
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)
ax.contour(x1_dash,x2_dash,f, cmap='plasma'); # Contour Plot, Standard Normal Dist. is perfectly circular. 
```
