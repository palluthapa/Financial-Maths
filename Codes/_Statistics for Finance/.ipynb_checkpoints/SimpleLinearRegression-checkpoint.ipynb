# Linear Regression
Lets load two sets of numeric data and creates a scatter plot. We will then fit a linear regression model to the data and add the regression line to the plot. Finally, we provide summaries of the model and its residuals.

```r
x=c(88.6,71.6,93.3,84.3,80.6,75.2,69.7,71.6,69.4,83.3,79.6,82.6,80.6,83.5,76.3)
y=c(20,16,19.8,18.4,17.1,15.5,14.7,15.7,15.4,16.3,15,17.2,16,17,14.4)
plot(x,y)
fit2=lm(y~x)  # fits a linear regression model.
lines(x,predict.lm(fit2))
summary(fit2)
summary(fit2$residuals)
plot(x,fit2$residuals)
abline(h=0)
qqnorm(rstandard(fit2))
qqline(rstandard(fit2))
```

## Skewness And Kurtosis
`Skewness`, also known as the adjusted Fisher-Pearson standardized moment coefficient $G_1$. Biased estimator of the
true skewness (i.e. $\mathbb{E}(G1)\neq$ skeweness).
- Symmetric distributions (i.e. symmetric pdf) have zero skewness.
- Positive skew (skewed to the right): long right tail (eg: income data).
- Negative skew (skewed to the left): long left tail (eg: GPA grades).

`Kurtosis` indicator of how fatter/thinner the tails of a distribution are wrt normal tails (kurtosis=3, excess kurtosis=0).
- Positive excess kurtosis (leptokurtic): fat tails in comparison to those of normal distribution.
- Negative excess kurtosis (platykurtic): thin tails in comparison to those of normal distribution.
- Zero excess kurtosis (mesokurtic): tails like the normal.
