```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
plt.rcParams['figure.figsize'] = [15,7]
```

```python
def get_call_option_data(ticker):
    tkr = yf.Ticker(ticker)
    option = tkr.option_chain('2024-09-20')  # Option chain for a specific expiration date.
    return option.calls    
```

```python
option = get_call_option_data('JPM')  # JP Morgan call option.
option['midPrice'] = (option['bid'] + option['ask']) / 2  # Average of Bid and Ask price.
option
```

```python
num_rows = option.shape[0]
num_options = num_rows
last_prices = np.zeros(num_options)
mid_prices = np.zeros(num_options)  
strikes = np.zeros(num_options)
for i in range(0, num_options):
    mid_prices[i] = option.iloc[i,14]
    strikes[i] = option.iloc[i,2]
    last_prices[i] = option.iloc[i,3]
    
assert strikes[0]==65.0, 'Check first row correct'
assert strikes[-1]==230.0, 'Check last row correct'

fig,[ax1,ax2]=plt.subplots(2,1)
ax1.plot( strikes, last_prices, c= '#d6d306')
ax1.set_xlabel('Strike')
ax1.set_ylabel('Last-Price')
ax1.set_title('JPMorgan Option Prices');

ax2.plot( strikes, mid_prices, c= '#7906d6' )
ax2.set_xlabel('Strike')
ax2.set_ylabel('Mid-Price');
 
```
