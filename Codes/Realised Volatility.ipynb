## Realised Volatility Cones
Actual volatility is instantaneous and cannot be measured, thus traders associate volatility with specific time periods such as daily, weekly or monthly periods in order to put volatility into context. Plotting Volatility cones allows us to have some sense of whether an option is overpriced or underpriced and helps in our option strategy selection.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
import mathfinance as mf
plt.rcParams['figure.figsize'] = [15,7]
```

```python
def read_logreturn_history(ticker, start, end, interval = '1d'):
    """ 
    Getting log Returns data from Yahoo finance.
                    
            Parameters:
                    ticker (string) : List of Stock tickers.
                    start (datetime) :  Start date of stock data to be extracted.
                    end (datetime) :  End date of stock data to be extracted.
                    interval (string) : Interval of stock data to be extracted (default is '1d').

            Returns:
                    log_return (double) : 2D Numpy array of stock log returns data.
    """
    # Create an empty DataFrame to store the Weekly Stock Price Data.
    stock_data = pd.DataFrame()
    
    # Fetch and append weekly data for each stock to the DataFrame.
    for tkr in ticker:
        stock = yf.download(tkr, start= start, end= end, interval= interval, progress=False);
        stock.reset_index(inplace=True)
        datetime = stock['Date'].to_numpy()[::-1]
        stock = stock[['Adj Close']]  # Selecting only Adj Close column
        stock.columns = [tkr]  # Rename column to stock symbol
        stock_data = pd.concat([stock_data, stock], axis=1)
    
    # Reset index to have a clean sequential index
    stock_data.reset_index(inplace=True)
    # Dealing with missing values.
    stock_data = stock_data.ffill()
    stock_data = stock_data.bfill()
    
    # Convert stock_data to a NumPy array
    price_history = stock_data.iloc[:, 1:].to_numpy()

    # Reverse the order of rows in the NumPy array to have the correct chronological order.
    # Array contains earliest date first and recent date later.
    price_history = np.log(price_history[::-1])
    
    # Computing Historical log returns.
    start_prices = price_history[:-1,:]    # All prices except last row.
    end_prices = price_history[1:,:]     # All prices except first row.
    log_return = (end_prices-start_prices)/start_prices
    return log_return, datetime

def rolling_vol(returns, window_size):
    rolling_volatility = np.zeros(len(returns) - window_size + 1)
    for i in range(len(returns) - window_size + 1):
        window_returns = returns[i:i+window_size]
        volatility = np.std(window_returns, ddof=1)*100 * np.sqrt(252)
        rolling_volatility[i] = volatility
    return rolling_volatility
```

```python
ticker = ['^FTSE']
start = '2000-01-01'
end = '2024-03-01'
interval = '1d'

log_returns, datetime = read_logreturn_history(ticker, start, end, interval = '1d')
rvol = rolling_vol(log_returns,20)

ax1 = plt.gca()
ax1.plot(datetime[1:], log_returns)
ax1.set_xlabel('$Date$')
ax1.set_ylabel('Log Returns');
ax1.set_title('FTSE 100 Log Returns', fontsize = 14);
```

```python
ax2 = plt.gca()
ax2.plot(datetime[20:], rvol)
ax2.set_xlabel('$Date$')
ax2.set_ylabel('Realised Daily Volatility');
ax2.set_title('FTSE 100 Volatility Clusters', fontsize = 14);
```
