### Probability Integral Transform
 If $X$ is a continuous random variable with cumulative distribution function $F_X(x)$, then the random variable $Y = F_X (X)$
 has a uniform distribution on $[0,1]$.

*Proof:*

let $Y = F_X (X)$ where $X$ is a continuous random variable.
$$ F_{Y}(y):=P({Y}\leq y)=P(F_X(X)\leq y)=P(X\leq F^{-1}_X(y))=F_X(F^{-1}_X(y))=y. \square$$ 

### Empirical Cumulative Distribution Function(ECDF)
If the original distribution of our data is unknown or the observations in a collected data sample do not fit any known probability distribution and cannot be easily forced into an existing distribution by data transformation or parameterization of the distribution function, then the empirical probability distribution should be used. It provides non-parametric estimate of a population’s cumulative distribution function (CDF). We can estimate our data’s CDF regardless of their underlying distribution. 

If $X_1, X_2, …, X_n$ is a sample of size $n$ from a population with CDF $F_X (x)$, Then the empirical CDF $F_n (x)$ is:
$$F_n (x) = \frac{1}{n} \sum_{i=1}^{n} \mathbb{1}_{(X_i \leq x )}$$

```python
# Fitting an Empirical cdf to a bimodal dataset.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
from matplotlib import cm
np.random.seed(777)
plt.rcParams['figure.figsize'] = [15,7]

def ecdf(data):
    """
    Compute the empirical cumulative distribution function (ECDF) 
    of a 1D array of data.
    """
    unique_values, counts = np.unique(data, return_counts=True)
    sorted_indices = np.argsort(unique_values)
    sorted_values = unique_values[sorted_indices]
    cumulative_counts = np.cumsum(counts[sorted_indices])
    probabilities = cumulative_counts / len(data)
    adjusted_probabilities = probabilities[sorted_indices]
    x = np.repeat(sorted_values, counts[sorted_indices])
    y = np.repeat(adjusted_probabilities, counts[sorted_indices])
    return (x,y)

# Generate a sample.
mu1 = 20;
sigma1 = 5;
sample1 = sigma1 * np.random.randn(400) + mu1;    # 400 Standard normal random numbers N(20,25).
mu2 = 40;
sigma2 = 5;
sample2 = sigma2 * np.random.randn(600) + mu2;    # 600 Standard normal random numbers N(40,25).
sample = np.hstack((sample1, sample2))

# Generating a histogram and Fitting a CDF.
cdf = ecdf(sample);
sns.set_style('darkgrid');
figure, [ax1, ax2] = plt.subplots(1,2);
ax1.title.set_text("Bimodel Distribution");
ax1.set_xlabel('X');
sns.histplot(sample, kde=True, ax = ax1, bins = 50, color = "#8eede8", stat = 'density');    # Histogram and KDE.
ax2.plot(cdf[0], cdf[1],linewidth = '3', color = "#8eede8");
ax2.axvline(38, c='#522d4f');    #60% of data is below 38.
ax2.set_xlabel('$X$');
ax2.set_ylabel('$F(X)$');
ax2.legend(["Empirical CDF"], loc ="lower right");
ax2.title.set_text("ECDF Of Bimodel Distribution");
```

### Copulas
Let $\textbf{X}=[x_1,x_2,...,x_n]$ be a random vector with each $x_i$ a random variable. Each $x_i$ has marginal CDF $F_i$ for $i \in [1,n]$. The probability integral transform yields $F_i (X_i) = U_i \sim U[0,1]$. 

So, $\textbf{u}=[u_1,u_2,...,u_n] = [F_1 (x_1), F_2 (x_2),...,F_n (x_n)]$

A Copula is a multivariate CDF of multiple random variables with uniform marginals. $C:[0,1]^n \to [0,1]$ is a copula with n-dimensional random vector(n-variables).

$$
\begin{split}
	C(\textbf{u}) & = C[F_1 (x_1), F_2 (x_2),...,F_n (x_n)] \\
	 & = C[u_1, u_2,...,u_n] \\
     & = P[U_1 \leq u_1, U_2 \leq u_2,...,U_n \leq u_n]
\end{split}
$$

$\textbf{Theorem: (Sklar's Theorem)}$ A random vector $X$ with CDF $F$ and univariate marginal CDF's $F_1,...,F_n, \exists$ a copula $C$ such that $$F(x_1,x_2,...,x_n)=C(F_1(x_1),...,F_n(x_n)$$ If $X$ is continuous then the copula $C$ is unique.

Sklar's Theorem allows us to write the joint CDF $F(\textbf{X})$ in terms of marginal CDFs $F_i$ and copula $C$. Conversely, if we have joint CDF $F$ and the marginals $F_1,...,F_n$, we can find copula via:
$$C(u_1, u_2,...,u_n) = F(F_1^{-1}(u_1),...,F_n^{-1}(u_n))$$
$\text{where,      } F_i^{-1}(t) = inf\{ s: F_i(s)\geq t\}$

We can express any CDF using a copula.

### Gaussian Copula
We will create bivariate Gaussian copula$(n=2)$ for two random variables.
$$ 
\begin{pmatrix}
Z_1 \\
Z_2 
\end{pmatrix}
\sim
\textbf{N}
\Biggl(
\begin{bmatrix}
0  \\
0 
\end{bmatrix}
,
\begin{bmatrix}
1 & \rho \\
\rho & 1 
\end{bmatrix}
\Biggl)
$$
So the Gaussian Copula is:
$$
\begin{split}
	C(u_1,u_2) & = P(U_1 \leq u_1, U_2 \leq u_2) \\
	 & = P(Z_1\leq \Phi^{-1}(u_1), Z_2\leq \Phi^{-1}(u_2)) \\
     & = \Phi_{2} \bigl(\Phi^{-1}(u_1), \Phi^{-1}(u_2)|\rho\bigl)
\end{split}
$$
where $\Phi$ is the CDF of Standard Normal distribution and $\Phi_{2}$ is the joint CDF parameterised by correlation factor $\rho$.

Lets write the steps for Gaussian Copula:
> 1. Generate marginals $Z_1 \sim N(0,1)$ and $Z_2 \sim N(0,1)$.
2. Use probability integral transform to convert gaussian marginals to uniform marginals.$$F_1(Z_1) = u_1\\ \text{&} F_2(Z_2) = u_2$$
3. Apply Gaussian Copula Formula $$ C(u_1,u_2)= \Phi_{2} \bigl(\Phi^{-1}(u_1), \Phi^{-1}(u_2)|\rho\bigl)$$ 
where, $$\Phi^{-1}= \text{Inverse of Standard Normal CDF}$$ and 
$$
\Phi_2= \text{CDF of Multivariate distribution}
\sim N \Biggl(
\begin{bmatrix}
0  \\
0 
\end{bmatrix}
,
\begin{bmatrix}
1 & \rho \\
\rho & 1 
\end{bmatrix}
\Biggl)
$$

```python
# Step 1: Generating the marginals Z1 AND Z2.
mu1 = 0;
sigma1 = 1;
mu2 = 0;
sigma2 = 1;

x = np.linspace(-2, 2, 100);
y = np.linspace(-2, 2, 100);

z1 = np.random.randn(1000);
z2 = np.random.randn(1000);

# Plot of two variables with bivariate and univariate graphs.
sns.set_style('darkgrid');
plot1 = sns.jointplot(x = z1, y = z2);
plot1.fig.suptitle("Joint Scatterplot");
plot1.fig.subplots_adjust(top=0.95);
plot1.set_axis_labels('$Z_1 \sim N(0,1)$', '$Z_2 \sim N(0,1)$', fontsize=11);

plot2 = sns.jointplot(ax=ax2, x = z1, y = z2, kind="kde", color = '#e01295')
plot2.fig.suptitle("Bivariate and Univariate KDEs");
plot2.fig.subplots_adjust(top=0.95);
plot2.set_axis_labels('$Z_1 \sim N(0,1)$', '$Z_2 \sim N(0,1)$', fontsize=11);

# rugplot plots marginal distributions by drawing ticks along the x and y axes
# and plot_marginals draw univariate plots on each marginal axes.
plot1.plot_marginals(sns.rugplot, color="#d6e012", height=.17, clip_on=False, alpha=0.5);
```

```python
# Step2: Use probability integral transform to convert gaussian marginals to uniform marginals.
u1 = stats.norm.cdf(z1);
u2 = stats.norm.cdf(z2);

plot3 = sns.jointplot(x = u1, y = u2);
plot3.fig.suptitle("Joint Scatterplot of Uniformly Distributed variables");
plot3.fig.subplots_adjust(top=0.95);
plot3.set_axis_labels('$U_1 \sim U[0,1]$', '$U_2 \sim N[0,1]$', fontsize=11);
plot3.plot_joint(sns.kdeplot, color = "#e01295", levels = 1);
plot3.plot_marginals(sns.rugplot, color="#d6e012", height=.17, clip_on = False, alpha = 0.5);
```

```python
# Step3: Apply Gaussian Copula Formula C(u1,u2) = Φ2[ Φ^-1(u1), Φ^-2(u2) ]

# Sorting makes the plot smoother. Setting last data value to 1 makes smoother plot.
u1=np.sort(u1);
u2=np.sort(u2);
u1[-1]=1;
u2[-1]=1;
# Φ^-1(u1), Φ^-2(u2)
x1=stats.norm.ppf(u1);  # Inverse of Standard Normal CDF [Φ^-1(u1)]
x2=stats.norm.ppf(u2);  # Inverse of Standard Normal CDF [Φ^-1(u2)]
# Calculating Φ2 which is a CDF of Multivariate Normal distribution.
mu_x = 0;
var_x = 1;
mu_y = 0;
var_y = 1;
rho = 0.8;
mu = [mu_x, mu_y];    # vector 2x1
sigma = [[var_x, rho], [rho, var_y]];    # Matrix 2x2

# CDF of Multivariate distribution Φ2
X, Y = np.meshgrid(x1,x2);
pos = np.dstack((X, Y));
rv = stats.multivariate_normal(mu, sigma);
phi2 = rv.cdf(pos);
Z = rv.pdf(pos);

# Plotting 3d graph
X, Y = np.meshgrid(u1,u2);
fig = plt.figure(figsize=(11,11));
ax = fig.add_subplot(projection='3d');
ax.plot_surface(X, Y, phi2, cmap = cm.spring, linewidth = 0);
ax.invert_yaxis();
ax.set_xlabel('X');
ax.set_ylabel('Y');
ax.set_zlabel('Z');
ax.set_title('Gaussian Copula of ($Z_1,Z_2$)', fontdict={'fontsize': 15});
plt.show();

# Plotting Contour lines
plt.contour(X, Y, Z);
plt.show();
plt.tight_layout();
```
